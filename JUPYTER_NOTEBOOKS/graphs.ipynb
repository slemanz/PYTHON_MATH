# Visualizing Data with Graphs

Weâ€™ll be using matplotlib to make graphs with Python.

```python
import matplotlib.pyplot as plt

x = [1, 2,3]
y = [2, 4, 6]

plt.plot(x,y)
plt.show()
```

## Marking Points on Your Graph

We have different markers, like: 'o', '*', 'x', and '+'. Using marker= includes a line connecting the points (this is the default). You can also make a graph that marks only the points that you specified, without any line connecting them, by omitting marker=.

```python
#plt.plot(x,y, marker="o")
plt.plot(x,y, "o")
plt.show()
```

## Comparing Multiple Lines of Data

We can plot multiple lines on a single graph.

```python
x = range(1,13)
y1 = [31.3, 37.3, 47.2, 51.0, 63.5, 71.3, 72.3, 72.7, 66.0, 57.0, 45.3, 31.1]
y2 = [40.9, 35.7, 43.1, 55.7, 63.1, 71.0, 77.9, 75.8, 66.6, 56.2, 51.9, 43.6]
y3 = [37.3, 40.9, 50.9, 54.8, 65.1, 71.0, 78.8, 76.7, 68.8, 58.0, 43.9, 41.5]

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

We have a problem, however, because we donâ€™t have any clue as to which color corresponds to which year. To fix this, we can use the function legend(),

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.legend(["y1", "y2", "y3"])
plt.show()
```

## Customizing Graphs

We can costumize our graph in many ways.

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.legend(["y1", "y2", "y3"])
plt.title("Graph's title")
plt.xlabel("X axis")
plt.ylabel("X axis")

plt.axis([0,12, 0, 100])

plt.show()
```
