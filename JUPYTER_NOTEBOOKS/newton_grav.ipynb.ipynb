# Newton’s Law of Universal Gravitation

According to Newton’s law of universal gravitation, a body of mass $m_1$ attracts another body of mass $m_2$ with an amount of force $F$ according to the formula:

$$ F=\dfrac{Gm_1m_2}{r^2} $$

Where $r$ is the distance between the two bodies and $G$ is the gravitational constant ($6.674 \cdot 10^{−11} \,N m^2 kg^{−2}$). Let's consider $m_1 = 0.5\,\textrm{kg}$ and $m_2 = 0.5\,\textrm{kg}$. Considering 19 distances between 100 to 1000 meters. We will have:

```python
import matplotlib.pyplot as plt

# 1. Draw the graph function
def draw_graph(x ,y):
    plt.plot(x, y, marker='o')
    plt.xlabel('Distance in meters')
    plt.ylabel('Gravitational force in newtons')
    plt.title('Gravitational force and distance')
    plt.show()

# 2. Compute the equation
def generate_F():
    # 2.1. generate values for r
    r = range(100, 1001, 50)
    
    # 2.2. list to store the calculated values to F
    F = []

    # 2.3. constant and masses
    G = 6.674*(10**(-11))
    m1 = 0.5
    m2 = 1.5

    # 2.4. calculate the force and append to F list
    for dist in r:
        force = (G*m1*m2)/(dist**2)
        F.append(force)

    # 2.5. call the draw graph function
    draw_graph(r, F)

# 3. Execute
generate_F()
```

As the distance ($r$) increases, the gravitational force decreases. With this kind of relationship, we say that the gravitational force is inversely proportional to the distance between the two bodies. Also, note that when the value of one of the two variables changes, the other variable won’t necessarily change by the same proportion. We refer to this as a __nonlinear relationship__.
